# Approximating a Gunshot Waveform with ReLU Networks

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Input
```

## Defining the base functions

```python
x_values = np.linspace(0, 1, 1000)
m = 2 # Precision parameter (5 Interpolation points)

# Define the activation function (vectorized for NumPy arrays)
def relu(x):
    return np.maximum(0, x)

# "tooth" function using ReLU operations
def tooth(x):
    return 2*relu(x) - 4*relu(x - 1 / 2) + 2*relu(x-1)

# Composition function
def g(x, s):
    if s == 0:
        return x
    return g(tooth(x), s - 1)

# Plotting the true function vs. approximation function
def plot(true_values, approx_values, x_values):
    plt.figure(figsize=(8, 6))
    plt.plot(x_values, true_values, label="Ground Truth", color="blue", linewidth=2)
    plt.plot(x_values, approx_values, label="Approximated values", color="red", linewidth=2)

    plt.xlabel("x values")
    plt.ylabel("y values")
    plt.title(f"Mean absolute error = {(np.mean(np.abs(true_values - approx_values))):.5f}")
    plt.legend()
    plt.grid()
    plt.show()
```

## Building the sawtooth function with ReLU networks

```python
plt.plot(x_values,g(x_values,1),label=r"$g$",c="red")
plt.plot(x_values,g(x_values,2),label=r"$g_2$",c="green")
plt.plot(x_values,g(x_values,3),label=r"$g_3$",c="goldenrod") # Saw tooth with 2**(m-1) teeth
plt.legend()
```

## Approximating the square function

```python
# Approximation function
def h_square(x, m):
    terms = np.array([g(x, s) / (2 ** (2 * s)) for s in range(1, m + 1)]) # 
    return x - np.sum(terms, axis=0)

h_square_values = h_square(x_values, m)
square_values = np.square(x_values)
plot(square_values,h_square_values,x_values)
```

## Approximating the multiplication function using squared sum formula

We create a multiplying function using expansion of squared sum, with a slightly different formula to accommodate for the fact that $x+y$ must be less than 1

```python
def h_mult(x,y,m):
    return (h_square((x+y)/2,m) - h_square(x/2,m) - h_square(y/2,m))*2
 
def h_power(x,m,k):
    if k == 0:
        return 1
    base = x
    for _ in range(k - 1):
        base = h_mult(base, x, m)
    return base
```

## Approximating the negative exponential function
We can also use Maclaurin series in order to approximate functions. Furthermore, we also have to redefine $m$ to get a better precision given the accumulation of approximation errors

```python
d = 10 # Big-O precision for Maclaurin series
m = 4 # 17 Interpolation points
```

```python
def h_mexp(x, m, d): 
    s = np.zeros_like(x)
    for i in range(0,d,2):
        s += h_mult(h_power(x,m,i),1/math.factorial(i),m)
    for i in range(1,d,2):
        s -= h_mult(h_power(x,m,i),1/math.factorial(i),m)
    return s

h_mexp_values = h_mexp(x_values,m,d)
mexp_values = np.exp(-x_values)

plot(mexp_values, h_mexp_values,x_values)
```

## Approximating the sine function

```python
def h_sin(x, m, d):
    s = np.zeros_like(x)
    for i in range(1,d,4):
        s += h_mult(h_power(x,m,i),1/math.factorial(i),m)
    for i in range(3,d,4):
        s -= h_mult(h_power(x,m,i),1/math.factorial(i),m)
    return s


h_sin_values = h_sin(x_values,m,d)
sin_values = np.sin(x_values)

plot(sin_values, h_sin_values,x_values)
```

Since the domain of definition of `h_square` is $[0,1]$, we cannot extend this approximation of $x \rightarrow sin(x)$ to $[0;\pi]$. Instead, we have to use Taylor Series centred around different points.

```python
x_values_sin = np.linspace(0, np.pi, 3000)

h_sin_values2 = h_sin(x_values_sin,m,d)
sin_values2 = np.sin(x_values_sin)

plot(sin_values2,h_sin_values2,x_values_sin)
```

## Approximating $e^{-x} \times \sin(x)$

```python
plot(np.multiply(sin_values,mexp_values), h_mult(h_sin_values,h_mexp_values,m),x_values)
```

## Theory Vs. Practice

Even though we built a neural network capable of approximating our signal with any error $\varepsilon>0$, the results in practice can be very different to theoretical results. This can notably be noticed in the $n$-ap problem, for which we can easily build a handcrafted solution but that reveals impossible to learn for a traditional neural network with random initial weights.

```python
# Data
X = np.array([[0],[.125],[.25],[.375],[.5],[.625],[.75],[.875]])
y = np.array([[0],[1],[0],[1],[0],[1],[0],[1]])
```

### Theory 

```python
plt.figure(figsize=(8, 6))
plt.plot(x_values.reshape((-1,1)),g(x_values,3),label="Network values")
plt.plot(X,y,'rx',label="Ground truth")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.title(r"Neural Network model for the $n$-ap problem")
plt.grid()
```

### Practice

```python
model = Sequential()
model.add(Input(shape=(1,)))
model.add(Dense(2, activation='relu'))
model.add(Dense(2, activation='relu'))
model.add(Dense(2, activation='relu'))
model.add(Dense(2, activation='relu'))
model.add(Dense(2, activation='relu'))
model.add(Dense(2, activation='relu'))
model.add(Dense(1, activation='relu'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=1000, batch_size=8,verbose=0)

scores = model.evaluate(X, y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

xx = x_values.reshape((-1,1))
plt.figure(figsize=(8, 6))
plt.plot(xx,model.predict(xx),label="Fitted values")
plt.plot(X,y,'rx',label="Ground truth")
plt.title(r"Neural Network model for the $n$-ap problem")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.grid()
```

## Network Architecture

We will study $e^{-x}\sin(x)$ again with lower precision and study its architecture to get an idea of what we should be aiming for in our practical application. We choose $m=3$ and $d=5$ to get a simpler architecture with a performance that is still satisfactory.

```python
m, d = 3, 5
```

Furthermore, we have to redefine the ReLU function in order to keep track of the amount of time it is being called.

```python
neurons = 0
# We also redefine tooth to keep track of the number of neurons 
def relu(x):
    global neurons
    neurons += 1
    return np.maximum(0, x)
```

```python
h_sin_mexp_values = h_mult(h_sin(x_values,m,d),h_mexp(x_values,m,d),m)
sin_mexp_values = np.multiply(sin_values,mexp_values)
plot(sin_mexp_values, h_sin_mexp_values,x_values)
print(f"Depth of the network:{neurons//3}, Width of the network: 3, Number of neurons: {neurons}.")
```
